```python
# Importing required Libraries

import pandas as pd
import requests
from bs4 import BeautifulSoup
import numpy as np
```

```python
headers = {"User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/106.0.0.0 Safari/537.36"}
webpage = requests.get("https://www.ambitionbox.com/list-of-companies?campaign=desktop_nav&page=1",headers= headers).text
```

```python
soup = BeautifulSoup(webpage,'lxml')
#print(soup.prettify())
```

```python
# Refind code to to scrap the data

CompanyData = soup.find_all("div", class_ = "company-content-wrapper")

CompanyName = []
Ratings = []
Reviews = []
CompanyType = []
CompanyHeadquater = []
CompanyAge = []
CompanyHeadcount = []


for i in CompanyData:
    CompanyName.append(i.find("h2").text.strip())
    Ratings.append(i.find("p", class_ = "rating").text.strip())
    Reviews.append(i.find("a", class_="review-count").text.strip())
    CompanyType.append(i.find_all("p", class_="infoEntity")[0].text.strip())
    CompanyHeadquater.append(i.find_all("p", class_="infoEntity")[1].text.strip())
    CompanyAge.append(i.find_all("p", class_="infoEntity")[2].text.strip())
    try:
        CompanyHeadcount.append(i.find_all("p", class_="infoEntity")[3].text.strip())
    except:
        CompanyHeadcount.append(np.nan)
    
# Now we need to create DataFrame

DataSet = {"CompanyName":CompanyName , "Ratings":Ratings, "Reviews":Reviews, "CompanyType": CompanyType, "CompanyHeadquater": CompanyHeadquater, "CompanyAge": CompanyAge, "CompanyHeadcount": CompanyHeadcount}
df = pd.DataFrame(DataSet)
df
```

```python
# let's do it for 30 pages
final = pd.DataFrame()
for j in range(1, 30):
    headers = {"User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/106.0.0.0 Safari/537.36"}
    webpage = requests.get(f"https://www.ambitionbox.com/list-of-companies?campaign=desktop_nav&page={j}",headers= headers).text
    soup = BeautifulSoup(webpage,'lxml')
    
    CompanyData = soup.find_all("div", class_ = "company-content-wrapper")

    CompanyName = []
    Ratings = []
    Reviews = []
    CompanyType = []
    CompanyHeadquater = []
    CompanyAge = []
    CompanyHeadcount = []


    for i in CompanyData:
        try:
            CompanyName.append(i.find("h2").text.strip())
        except:
            CompanyName.append(np.nan)
        try:
            Ratings.append(i.find("p", class_ = "rating").text.strip())
        except:
            Ratings.append(np.nan)
        try:
            Reviews.append(i.find("a", class_="review-count").text.strip())
        except:
            Reviews.append(np.nan)
        try:
            CompanyType.append(i.find_all("p", class_="infoEntity")[0].text.strip())
        except:
            CompanyType.append(np.nan)
        try:
            CompanyHeadquater.append(i.find_all("p", class_="infoEntity")[1].text.strip())
        except:
            CompanyHeadquater.append(np.nan)
        try:
            CompanyAge.append(i.find_all("p", class_="infoEntity")[2].text.strip())
        except:
            CompanyAge.append(np.nan)
        try:
            CompanyHeadcount.append(i.find_all("p", class_="infoEntity")[3].text.strip())
        except:
            CompanyHeadcount.append(np.nan)

    # Now we need to create DataFrame

    DataSet = {"CompanyName":CompanyName , "Ratings":Ratings, "Reviews":Reviews, "CompanyType": CompanyType, "CompanyHeadquater": CompanyHeadquater, "CompanyAge": CompanyAge, "CompanyHeadcount": CompanyHeadcount}
    df = pd.DataFrame(DataSet)
    final = final.append(df, ignore_index=True)
```
